```python
import pandas as pd
import pyanatomogram
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_csv('avg_agegaps.csv', index_col=0)
data = data.stack().rename_axis(['organ', 'condition']).to_frame('expression').reset_index()
data
```

```python
g = sns.FacetGrid(data, col='condition')
g.set_titles("{col_name}")
cmap = matplotlib.cm.Reds
norm = matplotlib.colors.Normalize(vmin=data['expression'].min(), vmax=data['expression'].max())
g.map(pyanatomogram.facetgrid_helper, 'organ', 'expression', name='homo_sapiens.male', cmap=cmap, norm=norm)
cbar_ax = g.fig.add_axes([0.92, 0.15, 0.02, 0.7])
cbar = plt.colorbar(
    matplotlib.cm.ScalarMappable(norm=norm, cmap=cmap), 
    cax=cbar_ax, 
    label='Relative expression',
    orientation='vertical'
)
cbar.ax.set_ylabel('z-scored age gap', rotation=270, labelpad=15)
plt.show()
```
